### Results for Question 2

```python
#Given sample code for question 2
def softmax(x):
    exp_x = np.exp(x)
    return exp_x/exp_x.sum()
#Example input vector
input_vector = np.array([10, 2.0, 0.1])
#Applying softmax
output_vector = softmax(input_vector)
#Displaying results
print("Input Vector: ")
print(input_vector)
print("\nSoftmax Output: ")
print(output_vector)
print("\nSum of Softmax Output: ")
print(output_vector.sum())
```

#### Mathematical Expression for softmax ($\sigma$) Function

Given $x = (x_{1}, x_{2}, ..., x_{K})$:
$$ \sigma(x)_{i} = \frac {e^{x_{i}}}{\sum_{j=1}^{K} e^{x_{j}}}$$

#### 2b

```python
#softmax using for loop
def softmax_loop(x):
    #Create array same length as x
    sigma = np.arange(0.0,len(x))
    #Sum variable
    exp_sum = 0
    for i in range(0, len(x)):
        #Find e^x_i
        exp_x = np.exp(x[i])
        #Add e^x_i to sum 
        exp_sum += exp_x
        #Store sigma(x)_i
        sigma[i] = exp_x/exp_sum
    return sigma
```

```python
#Check output using same input vector
output_vector2 = softmax_loop(input_vector)
print('Softmax_loop Output:\n', output_vector2)
#Compare values
print('Difference of Softmax and Softmax_loop functions for each value of x:\n', (output_vector2-output_vector))
```

#### (3a) Consider the first order linear differential equation 

$$ \frac{dy}{dt} = \sigma y$$

##### Solving for $y(t)$ by separation of variables


To separate $y$ terms and $t$ terms, multiply both sides by $\frac{dt}{y}$ 
$$ \frac{dy}{dt}\frac{dt}{y} = \sigma y\frac{dt}{y}$$

After simplifying:
$$ \frac {dy}{y} = \sigma dt$$

Integrate both sides:
$$ \int_{0}^{y}\frac{dy}{y} = \sigma\int_{0}^{t} dt$$

After integrating, the left hand side is:
$$ ln(y) - ln(0) = ln(y-0) = ln(y) $$
and the right hand side is:
$$\sigma(t-0) = \sigma t$$
thus,
$$ ln(y) = \sigma t $$

Exponentiating both sides to cancel out $ln$ term:
$$ e^{ln(y)} = e^{\sigma t} $$
and simplifying gives us this solution: 

$$ y(t) = e^{\sigma t}$$

#### (3b) Consider the second order linear differential equation 

$$ \frac{d^{2}y}{dt^{2}} - \sigma^{2}y = 0$$

##### Solving for $y(t)$ by using the characteristic equation
Assume our solution is of the form
$$ y(t) = e^{rt} $$ where r is some constant.

Differentiating this solution we get:
$$ \frac{d^{2}y}{dt^{2}} = r^{2}e^{rt}$$

$$ \frac{dy}{dt} = re^{rt}$$

Substituing these values: 
$$ \frac{d^{2}y}{dt^{2}} - \sigma^{2}y = r^{2}e^{rt} - \sigma^{2}e^{rt} = 0$$

Diving by $e^{rt}$ we get:
$$ r^{2} - \sigma^{2}  = 0 $$

Find factors $r_{1}$ and $r_{2}$
$$ r^{2} - \sigma^{2} = (r - \sigma)(r + \sigma) = 0 $$

$$ r_{1} = \sigma $$
$$r_{2} = -\sigma $$

Which gives us the two solutions:
$$ y_{1} = e^{\sigma t} $$
$$ y_{2} = e^{-\sigma t} $$

```python
#Define y(t) function in Eq(1) 
def y(t, sigma):
    return np.exp(sigma*t)

#Interval [0,3] with 150 data points
t = np.linspace(0,3,150)
# values of 0.1, 0.2, 0.3, and 0.4 
sigma = np.linspace(0.1,0.4,4)

#4 panel plot using for loop
for i in range(0,4):
    plt.subplot(2,2,i+1)
    plt.plot(t, y(t,sigma[i]), 'r')
    plt.title(f'$y(t) = e^{{\\sigma t}}$ for $\sigma = {sigma[i]:.1f}$') #sigma[2] prints as 0.30000000000000004 so round to 1 decimal
    plt.ylim(0,4)
    
plt.tight_layout() #tight_layout avoids plots overlapping onto titles
plt.show()
```
