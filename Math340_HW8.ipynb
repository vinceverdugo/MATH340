### Results for Question 4

Solve for 

$$\frac{\partial L}{\partial w_1} \,, \quad L = \frac{1}{2}(q - \hat{q})^2 $$


using symbolic computing

```python
#Define variables as symbols
u, v, w1, w2, w3, b1, b2, x, a, z, q, qhat, L = sp.symbols('u, v, w1, w2, w3, b1, b2, x, a, z, q, qhat, L')

#Define the functions x, a, z, q^hat and L 
x = w1*u + w2*v + b1
a = x
z = w3*a + b2
qhat = z
L = 0.5*(q-qhat)**2

#Solve dL/dw_1 and substitute the numerical value of all variables 
dL_dw1 = sp.diff(L, w1)
dL_dw1_sol = dL_dw1.subs({u: 0.5, v: 0.1, w1: 0.4, w2: 0.3, w3: 0.7,
                         b1: 0.1, b2: 0.2, q: 1})

print('Numerical value of dL/dw_1 =', dL_dw1_sol)
```

Comparing our results to the results from Homework 7, our result is fairly similar. The largest difference between the results using symbolic computing and manual calculation is that with symbolic computing, our numerical value of $\frac{\partial L}{\partial w_1}$ is negative while our value from from HW7 (0.19914999999999997) was positive. Other than that, our value is the same when rounded to the 5th decimal place.

### Results for Question 5

```python
import numpy as np
import matplotlib.pyplot as plt

#Define parameters
u0 = 0.002
eps = 1e-5
v0 = u0 + eps
sig = 2

#Define delta-t and time interval
dt = 0.001
t = np.arange(0, 10, dt)

U = (u0*np.exp(sig*t))/(1-u0+u0*np.exp(sig*t))
V = (v0*np.exp(sig*t))/(1-v0+v0*np.exp(sig*t))
```

```python
#Compute functions 
#Using U to represent Z
a1 = (1/dt)*(np.log(U[1:]/U[:-1]))
a2 = sig*(1-U)

b1 = (1/dt)*(np.log((U[1:]-V[1:])/(U[:-1]-V[:-1])))
b2 = sig*(1-2*U)
```

```python
#Generate plots

#Plot of U
plt.subplot(2,2,1)
plt.plot(t, U, 'k')
plt.title(r'Plot of U = $\frac{u_0 e^{\sigma t}}{1 - u_0 + u_0 e^{\sigma t}}$')

#Plot of differences between U and V
plt.subplot(2,2,2)
plt.plot(t, (U-V), 'r--', label='U - V')
plt.title(r'Difference of U and V = $\frac{v_0 e^{\sigma t}}{1 - v_0 + v_0 e^{\sigma t}}$')
plt.legend()

#Plot of a1 and a2
plt.subplot(2,2,3)
plt.plot(t[1:], a1, linewidth=2, label='a1')
plt.plot(t, a2, '--', linewidth=2, label='a2')
plt.title('Growth Rate Solutions a1 and a2')
plt.legend()

#Plot of b1 and b2
plt.subplot(2,2,4)
plt.plot(t[1:], b1[0:], linewidth=2, label='b1')
plt.plot(t, b2, '--', linewidth=2, label='b2')
plt.title('Growth Rate Solutions b1 and b2')
plt.legend()

plt.tight_layout()
plt.show()
```
