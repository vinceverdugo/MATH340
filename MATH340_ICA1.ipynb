#### 1. Indentation
Python uses indentation to indicate the order or statements or instructions in code. To do this, they impliment white space as opposed to using braces. For example, in the code

<code>for x in (r):
    f_list.append(x)
print(f_list)
</code>

the indentation before <code>f_list.append(x)</code> indicates that it belongs to the for loop and that will be the statment executed by the loop. Once the loop is over, the code will then move down to the <code>print(f_list)</code> statement and that will be executed next. 

#### 2. Concatenation of adjacent string literals
In Python, two string literals on parallel lines can be added together by using a <code>\\</code>

For example:

<code>str = 'Hello' \\
       'World'
</code>

is equivalent to 

<code>str = 'Hello World'
</code>

#### 3. Arithmetic
Python arithmetic includes operators such as <code>+</code> for addition, <code>-</code> for subtraction, <code>*</code> for multiplication, <code>/</code> and <code>//</code> for division, <code>%</code> for modulus, and <code>**</code> for exponentiation

The difference between <code>/</code> and <code>//</code>, is that <code>/</code> stands for "true division" meaning the quotient of two integers will be returned as a float, while <code>//</code> as a quotient will be returned as an integer

#### 4. Comparison Operators
Python uses the comparison operators <code>=</code>, <code>!=</code>, <code><</code>, <code>></code>, <code><=</code>, <code>>=</code>, <code>is</code>, <code>is not</code>, <code>in</code>, <code>not in</code>
    
These operators can be used on data types such as numbers, strings, sequences, and maps. Additionally they can be chained to evaluate multiple terms. When chained, they will be evaluated in order from left to right. 

#### 5. Class and static methods
In Python, class and static methods can be created by <code>@classmethod</code> and <code>@staticmethod</code>. While the static method does not require an argument to be passed, the class method does.

```python
import math
import numpy as np

sigma = 40
dt = 0.1
amp = 1.0

for n in range(1,200):
    r1 = np.float64(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float64(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float64(r1*math.exp(arg))
    print ('%d times exp(-sigma dt) and exp (sigma dt): %.16f'%(n,r1))
```

What this code is doing is trying to find at what value of $n$ does the function no longer evaluate to 1. As we see, the exponential function should evaluate to 1 as we reach infinity, but as we increase to $n=187$, we no longer see that and instead, our function evaluates to 0

```python
import math
import numpy as np

sigma = 40
dt = 0.1
amp = 1.0

for n in range(1,200):
    r1 = np.float32(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float32(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float32(r1*math.exp(arg))
    print ('%d times exp(-sigma dt) and exp (sigma dt): %.16f'%(n,r1))
```

By changing our value of r1 from a float with 64 bits of memry to a float with 32 bits, we see that the value of $n$ such that our function evaluates to 0 goes from 187 to just 26. This loss of precision in our value of r1 results in a function that decreases much faster than before. 

```python
import matplotlib.pyplot as plt

sigma = 40
dt = 0.1
amp = 1.0
#Initialize arrays to store values of n and r1 from float64 version and float32 version
n_val64 = []
r_val64 = []
n_val32 = []
r_val32 = []

for n in range(1,28):
    r1 = np.float64(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float64(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float64(r1*math.exp(arg))
    
    n_val64.append(n)
    r_val64.append(r1)
    
for n in range(1,28):
    r1 = np.float32(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float32(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float32(r1*math.exp(arg))
    
    n_val32.append(n)
    r_val32.append(r1)
    
plt.subplot(1,2,1)
plt.plot(n_val64, r_val64)

plt.subplot(1,2,2)
plt.plot(n_val32, r_val32)
```

Given the following data: The principal amount (i.e., the total loan) is P = $420,000, the annual interest rate is 6.75%(This needs to be converted into the monthly rate (6.75/12)%), and the loan is to be paid off in 15 years (equivalent to 180 months). Use the above-derived formula and these data to compute the monthly mortgage payment x by Python. The answer should be an amount of money per month.

(b)

Without further instructions, ChatGPT was able to identify which formula would be used to solve the question. It then specified which data had already been given and assigned it to variables. It then created code (below) which after transfering to Jupyter Notebook and running provided the answer

```python
# Given data
P = 420000  # Principal loan amount
annual_rate = 6.75  # Annual interest rate in percent
years = 15  # Loan duration in years

# Calculating monthly interest rate and number of payments
monthly_rate = (annual_rate / 100) / 12  # Convert annual rate to monthly rate (decimal)
n = years * 12  # Total number of monthly payments

# Monthly payment formula
monthly_payment = P * (monthly_rate * (1 + monthly_rate)**n) / ((1 + monthly_rate)**n - 1)

# Display the result
print(f"Monthly mortgage payment: ${monthly_payment:.2f} per month")
```
