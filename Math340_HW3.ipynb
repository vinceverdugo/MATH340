### Results for Question 1 (within a Markdown cell)

```python
#Given code for question 1
import math
inc = 3

for n in range(1,60,inc):
    r = 2.0
    for i in range(n):
        r = math.sqrt(r)
    for i in range(n):
        r = r**2
    print ('%d times sqrt and **2: %.16f'%(n,r))
```

What the code above is doing is taking $r=2$ and performing $\sqrt r$ $n$ times, then $r^2$ $n$ times. In theory, because we are performing opposite functions for the same number of times, they should cancel out and at the end 'r' should equal 2. In reality, we see that when we compile, we lose accuracy due to rounding. Starting at $n=1$, we are accurate up to 16 decimal places. As we increment by 3 each iteration, we see that as n increases, our end result is further away from 2. This is especially true when we see that beginning when $n=52$, we are left with $r=1.0000000000000000$. 

### Results for Question 2 (within a Markdown cell)

```python
#Import packages
import matplotlib.pyplot as plt
import numpy as np

#Defining functions from https://en.wikipedia.org/wiki/Activation_function#Table_of_activation_functions
def sig(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def ReLU(x): #Will return the higher value between (0,x), will return 0 when x <= 0, x when x>0
    return np.maximum(0, x)
```

```python
#Define function ranges
#tanh and ReLU functions will have same range
r1 = np.linspace(0,5,100)
r2 = np.linspace(-5,5,100)

#Call subplot function, create title for graph, add grid
plt.subplot(1,3,1)
title1 = r'Graph of $\sigma$ = $\frac{1}{1+e^{-x}}$'
plt.title(title1, size=10)
plt.plot(r1, sig(r1))
plt.grid()

plt.subplot(1,3,2)
title2 = r'Graph of $tanh$ = $\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$'
plt.title(title2, size=10)
plt.plot(r2, tanh(r2))
plt.grid()

plt.subplot(1,3,3)
title3 = r'Graph of $x^{+}$ = $max(0,x)$'
plt.title(title3, size=10)
plt.plot(r2, ReLU(r2))
plt.grid()
#plt.tight_layout() spaces out graphs more
plt.tight_layout()
plt.show()
```
