```python
a = 1.0
while a > 0.0:
    a = a/2.0
print(a)
```

```python
a = 1.0
while a + 1 > 1.0:
    a = a/2.0
print(a)
```

```python
a = np.float64(1.0)
while a + 1 > 1.0:
    a = a/2.0
print(a)
```

```python
b = np.float32(1.0)
while b + 1 > 1.0:
    b = b/2.0
print(b)
```

The last three programs all output the same number, however the first progam ouputs 0. This happens because we are using 0.0 as the threshold instead adding 1 to a and evaluating it greater than 1.0. We see this occuring because we are trying to evaluate a value that is incredibly small being greater than 0.0, which due to floating-point arithmetic and precision return 0.0 because we cannot get precise enough. We also see our next three programs returing the same values because they are all evaluated at or below the same precision. Because we are dividing by 1.0 in all the programs, even though a or b have higher precision, we do not see that in our return value. If we define two more variables c and d to be higher precision, we should see a difference in our return value
